# Erro Absoluto Médio

O erro absoluto médio (MAE) é uma métrica de avaliação utilizada em modelos de Regressão. Essa métrica aponta a média da distancia absoluta entre os valores reais e os valores previstos pelo modelo, isso é, quanto em média o modelo erra, sem informar se esse valor é para mais ou para menos.

**O que será abordado**

- Imports
- Dados de teste
- Divisão dos dados
- Modelo
- Erro Absoluto Médio

> AVISO: Temas relacionados à criação de modelos de Regressão não serão aprofundados nesse documento, porém, explicações sobre esse processo podem ser encontradas na pasta [Modelos/Regresssao](../Modelos/Regressao/README.md)

# Imports

```python
# Import da Biblioteca numpy
import numpy as np

# Import da biblioteca matplotlib
import matplotlib.pyplot as plt

# Import da função train_test_split -> Usada para dividir os dados usados para treinar o modelo dos dados utilizados para avaliar o modelo
from sklearn.model_selection import train_test_split

# Import da Classe LinearRegression -> Usada para criar o modelo que será avaliado
from sklearn.linear_model import LinearRegression

# Import da função mean_absolute_error -> Utilizada para dir o erro absoluto médio
from sklearn.metrics import mean_absolute_error
```

# Dados de teste

Nessa seção será desenvolvido um código com numpy que visa criar um conjunto de dados que possa ser usado para o teste de um modelo de Regressão Linear simples.

```python
# Define uma seed para a criação de valores aleátorios -> Garante que os dados gerados serão os mesmos e todas as execuções, possibilitando analises reprodutiveis
np.random.seed(42)

# Criação do conjunto de Feature
x = np.random.rand(100, 1) * 10

# Criação do conjunto de Targets
y = 2.5 * x + np.random.randn(100, 1) * 2 

# Plot dos dados gerados -> Exibe um gráfico com cada par de entrada e saída
plt.scatter(x, y)
plt.show()
```

Explicação dos dados:
- x -> A função `np.random.rand(100, 1)` cria uma matriz de 100 linhas e uma coluna com valores de 0 a 1. Multiplicar os valores por 10 implica que esses números são distribuídos de 0 a 10

- y -> a expressão `2.5 * x` faz com que a relação entre x e y seja linear, ideal para o teste de uma IA desse tipo. Já a expressão `+ np.random.randn(100, 1) * 2` faz com que um ruido seja adicionado aos dados, o que é comum em dados reais. 

Nota-se que, no conjunto de dados criado, os valores y (que representam os targets) tendem a aumentar de acordo com os valores x (que representam as features), sendo assim, um cenário para a possível aplicação de um modelo de Regressão Linear simples.

# Divisão dos dados

Conforme abordado no tópico sobre a [validação](../Validacao/README.md) de um modelo, é necessário separar os dados que serão utilizados para treinar o modelo dos dados utilizados para realizar testes de desempenho.

Portanto, nessa sessão será desenvolvido um código que possa separar os dados de treino e de teste do modelo, para que eventualmente os dados de teste possam ser usados para medir a qualidade do modelo.

```python
# realiza a divisão dos dados de treino e dos dados de teste
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=42)
```

Com isso foram criados 4 datasets sendo:

- **x_train** = features de treino
- **y_train** = targets de treino
- **x_test** = features de teste
- **y_test** = targets de teste

# Modelo

O Erro Absoluto Médio é uma métrica de avaliação utilizada para medir o desempenho de modelos de regressão. 

Para demonstrar o uso dessa métrica será criado um modelo de regressão linear simples utilizando os dados desenvolvidos até agora.

```python
# Criação de um modelo de regressão linear com o nome 'model'
model = LinearRegression()

# Treino do Modelo
model.fit(x_train, y_train)
```

A execução do código acima cria um modelo de regressão linear simples chamado 'model'. Esse modelo pode ser avaliado posteriormente com a métrica de Erro Absoluto Médio.

# Erro Absoluto Médio

A avaliação com a métrica de Erro Absoluto Médio pode ser realizada com a função `mean_absolute_error()` da biblioteta scikit-learn. Essa função recebe dois parâmetros principais, sendo:

- **y_true** = Valores reais de teste   
- **y_pred** = Valores previstos pelo modelo

Veja um exemplo de seu uso abaixo:

```python
# Realiza previsões para medir o desempenho do modelo
pred = model.predict(x_test)

# obtém dados de avaliação do modelo 
mae = mean_absolute_error(y_test, pred)

# Exibe os dados de Erro Absoluto Médio
print(f'O Erro Absoluto Médio do modelo é: {mae:.2f}')
```

No código acima foi utilizado o método `.predict()` do modelo criado, o retorno dessa função foi armazenado na variável 'pred' e contém as previsões realizadas pelo modelo.

Após realizar as previsões do modelo é possível avaliar o resultado através da função `mean_absolute_error()`, o retorno da função é o erro absoluto médio, que foi armazenado na variável mae.

O resultado obtido mostra um valor de 1.18 indicando que, em média o resultado fornecido pelo modelo possui uma distancia média de 1.18 do resultado real.
